```python
import pandas as pd

url = "https://raw.githubusercontent.com/Davis119/Posttest2_KB_Davis119/main/Penyebab%20Kematian%20di%20Indonesia%20yang%20Dilaporkan%20-%20Clean.csv"
dataset = pd.read_csv(url)

#Note ini untuk versi file yang Clean
#Note Posttestnya ada di bawah
#Mohon maaf apabila nomornya tidak ber-urut

dataset.head()
```

```python
#Ketentuan Nomor5
# Tampilkan jumlah label dari semua attribute yang bernilai object


# Memilih atribut dengan tipe data objek (string)
objek_atribut = dataset.select_dtypes('object')

# Inisialisasi dictionary untuk menyimpan jumlah label masing-masing atribut objek
hitung_label = {}

# Loop melalui setiap atribut objek dan hitung jumlah label unik
for kolom in objek_atribut.columns:
    hitung_label[kolom] = len(objek_atribut[kolom].unique())

# Menampilkan jumlah label dari Atribut yang Bernilai Object
print("Jumlah Label dari Atribut yang Bernilai Object:")
for atribut, jumlah in hitung_label.items():
    print(f"{atribut}: {jumlah} label")
```

```python
#Ketentuan Nomor4
# Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM 
#akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.
#metode dataset menggunakan slice .iloc dengan indeks label
metode1 = dataset.iloc[119:, 4:]
print(metode1)
```

```python
#Ketentuan Nomor4
# Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM 
#akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.
# metode dataset menggunakan slice .loc dengan indeks label
metode2 = dataset.loc[119:, dataset.columns[4]:]
print(metode2)
```

```python
#Ketentuan Nomor 3
#Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. 
#Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. 
#Tampilkan isi dari dictionary tersebut sebagus mungkin.


# berfungsi untuk menampilkan jumlah record
jumlah_record = len(dataset['Total Deaths'])

# berfungsi untuk menampilkan jumlah atribut
jumlah_attribute = len(dataset.columns)

#berfungsi untuk menampilkan nama attribute
#Dengan menggunakan fungsi list kita mengonversi objek dataset.
#columns yang awalnya dalam format Pandas menjadi list Python biasa. 
#dengan kata lain dari codingan list dibawah ini adalah kita mendapatkan daftar nama2 kolom dari DataFrame dalam bentuk list Python
nama_attribute = list(dataset.columns)

# Informasi jumlah attribute dengan tipe angka
jumlah_attribute_angka = len(dataset.select_dtypes(include=['int64']).columns)

# Menyimpan informasi dalam dictionary
informasi_dict = {
    "Jumlah Record": jumlah_record,
    "Jumlah Attribute": jumlah_attribute,
    "Nama Attribute": nama_attribute,
    "Jumlah Attribute dengan Tipe Angka": jumlah_attribute_angka
}

# Menampilkan isi dictionary
print("Informasi Dataset:")
for key, value in informasi_dict.items():
    print(f"{key}: {value}")
```

```python
#Ketentuan Nomor 1
#Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.
#note standar deviasi adalah sejauh mana angka2 dalam antar-set data kita berbeda dari rata-ratanya.
#note Q1 bagi data sehingga 25% data berada di bawahnya.
#note Q2 membagi data menjadi dua bagian yang sama besarnya , 50% data berada di bawah Q2 dan 50% berada di atasnya.
#note Q3 membagi data menjadi dua bagian  tetapi untuk Q3 ini memisahkan 25% data yang tertinggi atau dengan kata lain, 75% data berada di bawah Q3.

# Memilih atribut dengan tipe data angka
atribut_angka = dataset.select_dtypes(include=['int64'])

# Mendapatkan statistik dasar
statistik = atribut_angka.describe()


rata_rata = statistik.loc['mean']
standar_deviasi = statistik.loc['std']
nilai_minimum = statistik.loc['min']
nilai_maksimum = statistik.loc['max']
quartil1 = statistik.loc['25%']  # Q1
quartil2 = statistik.loc['50%']  # Q2
quartil3 = statistik.loc['75%']  # Q3


print("Rata-rata:")
print(rata_rata)
print("\nStandar Deviasi:")
print(standar_deviasi)
print("\nNilai Minimum:")
print(nilai_minimum)
print("\nNilai Maksimum:")
print(nilai_maksimum)
print("\nQuartile 1 (Q1):")
print(quartil1)
print("\nQuartile 2 (Q2, Median):")
print(quartil2)
print("\nQuartile 3 (Q3):")
print(quartil3)
```
